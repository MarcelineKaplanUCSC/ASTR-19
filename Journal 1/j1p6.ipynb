Import the libraries that we will need

```python
import numpy as np
from astropy.table import Table
```

Define function sin(x)
Input: x, a data array of floating point type input values
Output: A data array of floating point type value, where each is equal to the sine of the corresponding input

```python
def sin(x):
    return np.sin(x)
```

Define function cos(x)
Input: x, a data array of floating point type input values
Output: A data array of floating point type value, where each is equal to the cosine of the corresponding input

```python
def cos(x):
    return np.cos(x)
```

Tabulate a data array of the 1000 equally spaced points between 0 and 2PI

```python
x = np.linspace(0.0,2.0 * np.pi, 1000)
```

Create a data table consisting of columns x, sin(x), and cos(x), all labeled accordingly
Then, print the first ten elements of the data table

```python
dataTable = Table([x,sin(x),cos(x)],names=['x','sin(x)','cos(x)'])
print(dataTable[0:10])
```
